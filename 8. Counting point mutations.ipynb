```python
def findpointmutations(dna1, dna2):
    point_mutation_count = 0
    lengthofstring = len(dna1)
    for i in range(0, lengthofstring):
        if dna1[i] == dna2[i]:
            point_mutation_count += 0
        else:
            point_mutation_count += 1
    print point_mutation_count
```
