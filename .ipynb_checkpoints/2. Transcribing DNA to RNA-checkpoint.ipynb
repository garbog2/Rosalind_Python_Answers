```python
def dnatorna(dna):
    dnafile = open(dna, 'r')
    dna = dnafile.read()
    rna = dna.replace('T', 'U')
    answer = open('rosalind_rna_answer.txt', 'w+')
    print rna
    answer.write(rna)
    answer.close() 
    dnafile.close()
```
